```python
a1, a2, a3=map(int,input().split())
inp = []
for i in range(a3):
    a = input().split()
    inp.append([int(i) for i in a])
#继续跑
rec = [-1 for i in range(a3)]
rec[-1] = a1-inp[-1][0]
#预期第x个障碍
for i in range(a3):
    k=a3-i-1
    if k==a3-1:#从最后一个障碍前进的预期用时
        rec[k]=a1-inp[-1][0]
    else:
        #从第k个障碍前进，预期用时=距离耗时+成功率*（k+1预期）+失败率*（k+1预期+倒地时间）
        rec[k]=inp[k+1][0]-inp[k][0]+inp[k+1][1]*rec[k+1]+(1-inp[k+1][1])*(rec[k+1]+inp[k+1][2])
print(rec)
```

```python
def f3x_1(inpu):
    maxi = inpu
    while inpu!=1:
        if inpu%2==0:
            inpu/=2
        else:
            inpu = 3*inpu+1
            maxi = max(inpu,maxi)
    return maxi
```

```python
def test(upper):
    rg = range(1,upper)
    lis = [f3x_1(i) for i in rg]
    from matplotlib import pyplot as plt
    import numpy as np
    print(np.argmax(lis),max(lis))
    plt.plot(rg,lis)
    plt.grid(alpha=0.5)
    plt.show()
```

```python
test(300000)
```
